```python
    # Casting column
    from pyspark.sql.functions import col, cast, expr

    emp_casted = emp.select("employee_id", "name", "age", col("salary").cast("double"))
    emp_casted.show()
    emp_casted.printSchema()
```

```python
# Add Column
# Solution 1
emp_tax_1 = emp_casted.select('employee_id', 'name', 'age', 'salary', (col('salary') * 0.2).alias('tax_salary'))
emp_tax_1.show()
emp_tax_1.printSchema()
```

```python
# Solution 2
emp_tax_2 = emp_casted.withColumn("tax_salary", col("salary") * 0.2)
emp_tax_2.printSchema()
emp_tax_2.show()
```

```python
# Solution 3
emp_tax_3 = emp_casted.selectExpr(
    "employee_id as emp_id",
    "name",
    "age",
    "salary",
    "salary * 0.2 as tax_salary"
)
emp_tax_3.printSchema()
emp_tax_3.show()
```

```python
from pyspark.sql.functions import lit

emp_new_cols = emp_tax_1.withColumn('ColumnOne', lit(1)).withColumn('ColumnTwo', lit("two"))
emp_new_cols.printSchema()
emp_new_cols.show()
```

```python
# Rename column
emp_cols_renamed = emp_new_cols.withColumnRenamed('ColumnTwo', 'Column Two')
emp_cols_renamed.show()
```

```python
# Drop column
emp_dropped = emp_cols_renamed.drop('Column Two')
emp_dropped.show()
```

```python
# filter tax
emp_filtered_tax = emp_dropped.where('tax_salary > 10000')
emp_filtered_tax.show()
```

# Session 4: String and Date

```python
from pyspark.sql.functions import when, expr, col
```

```python
# Case - When
# Solution 1
emp_gender_fixed_1 = emp.withColumn('new_gender',
                                    when(col('gender') == 'Male', 'M')
                                    .when(col('gender') == 'Female', 'F')
                                    .otherwise(None))

emp_gender_fixed_2 = emp.withColumn('new_gender',
                                    expr('case when gender = "Male" then "M" when gender = "Female" then "F" else null end'))
```

```python
emp_gender_fixed_1.show()
```

```python
emp_gender_fixed_2.show()
```

```python
# Replace in String
from pyspark.sql.functions import regexp_replace

emp_name_fixed = emp.withColumn("new_name", regexp_replace(col('name'), 'J', 'Z'))
emp_name_fixed.show()
```

```python
from pyspark.sql.functions import to_date

emp_date_fixed = emp_name_fixed.withColumn('hire_date', to_date(col('hire_date'), 'yyyy-MM-dd'))
```

```python
emp_date_fixed.show()
```

```python
emp_date_fixed.printSchema()
```

```python
# Add date columns
from pyspark.sql.functions import current_date, current_timestamp

emp_dated = emp_date_fixed.withColumn("date_now", current_date()).withColumn('timestamp_now', current_timestamp())
```

```python
emp_dated.show()
```

```python
# Drop null
emp_1 = emp_dated.na.drop()
emp_1.show()
```

```python
# Fix null value
from pyspark.sql.functions import coalesce, lit
emp_null_df = emp_gender_fixed_2.withColumn('new_gender', coalesce(col('new_gender'), lit('O')))
emp_null_df.show()
emp_null_df.printSchema()
```

```python
emp_final = emp_null_df.drop('gender').withColumnRenamed('new_gender', 'gender')
emp_final.show()
```

```python
# write data to csv
emp_final.write.format('csv').save("data/output/3/emp.csv")
```

```python
# convert date to string and extract date information
from pyspark.sql.functions import date_format

emp_final = emp_final.withColumn('dateString', date_format(col('hire_date'), 'dd/MM/yyyy'))
emp_final.show()
```

```python
# extract information
emp_final = emp_final.withColumn('date_Year', date_format(col('hire_date'), 'yyyy'))
```

```python
emp_final.show()
```

```python
# sort the emp dataframe base on desc salary
from pyspark.sql.functions import desc, asc, col

emp_sorted = emp.orderBy(col("salary").asc())
emp_sorted.show()
```

```python
# aggregation count
from pyspark.sql.functions import count

emp_count = emp.groupBy('department_id').agg(count('employee_id'))
emp_count.show()
```
